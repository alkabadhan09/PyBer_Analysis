```python
%matplotlib inline
```

```python
# import dependencies

import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot

plt.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()

fig, ax = plt.subplots()

ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()

ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# create the plot

plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# create the plot

plt.bar(x_axis, y_axis)
```

```python
# create the plot
plt.bar(x_axis, y_axis, color = "green", label = 'Boston')

# create the lables for x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

# create a title
plt.title("PyBer Fare by Month")

# add the legend
plt.legend()
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
plt.gca().invert_yaxis()
ax.barh(x_axis, y_axis, color = "Cyan", label = 'Chicago')

ax.legend()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")

```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()

                               
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.subplots(figsize=(8,8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, labels=x_axis, explode = explode_values, autopct = '%.1f%%', colors = colors)
```

```python
fig, ax = plt.subplots(figsize=(8,8))

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode = [0,0,0.1,0,0,0,0.2,0,0,0,0,0]
ax.pie(y_axis, labels=x_axis , colors = colors, explode = explode, autopct = '%.1f%%')

plt.show()
```
